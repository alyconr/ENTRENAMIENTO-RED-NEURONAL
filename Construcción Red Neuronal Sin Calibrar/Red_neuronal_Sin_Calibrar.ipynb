# 1. Librerías

```python
#Importamos la librería de pandas
import pandas as pd

#Librería para manipular los servicios de AWS
import boto3

#Utilitario para manipular el sistema operativo
import os
```

# 2. Lectura del archivos

```python
#Bucket en donde se encuentran los archivos
#IMPORTANTE: REEMPLAZAR "XXX" POR TUS INICIALES
bucket = "datasetsbdajac"
```

```python
#Ruta de los archivos
rutaArchivos = "data/insurance_dataset/"
```

```python
#Nos conectamos al cliente de S3
s3 = boto3.client("s3")
```

```python
#Realizamos la consulta para listar los archivos de un directorio
respuesta = s3.list_objects(
    Bucket = bucket,
    Prefix = rutaArchivos
)
```

```python
#Obtenemos los archivos desde la respuesta
listaDeArchivos = respuesta.get("Contents", [])
```

```python
#Verificamos
listaDeArchivos
```

```python
#Lista para almacenar los DataFrames de cada archivo
dataframes = []
```

```python
#Itera sobre todos los archivos en el directorio
for archivo in listaDeArchivos:
    #Obtenemos el nombre del archivo
    rutaArchivo = archivo["Key"]

    #Nos quedamos sólo con los CSV
    if rutaArchivo.endswith('.csv'):
        #Concatenamos el nombre del archivo en la ruta
        rutaCompleta = f"s3://{bucket}/{rutaArchivo}"

        #Imprimimos la ruta del archivo
        print(rutaCompleta)

        #Leemos el archivo
        df = pd.read_csv(rutaCompleta, header = None)

        #Lo agregamos a la lista
        dataframes.append(df)
```

```python
#Concatenamos todos los dataframes
dfpDataset = pd.concat(dataframes, ignore_index = True)
```

```python
dfpDataset
```

# 3. Definición de features

```python
#Seleccionamos las columnas features, de la 1 en adelante
dfpFeatures = dfpDataset[dfpDataset.columns[1:]]

#Verificamos
dfpFeatures
```

# 4. Definición de labels

```python
#Seleccionamos la columna label, la primera
dfpLabels = dfpDataset[dfpDataset.columns[0]]

#Verificamos
dfpLabels
```

# 5. Split de registros para entrenar (dfTrain) y validar (dfTest)

```python
#Utilitario para dividir los datos en entrenamiento y validación
from sklearn.model_selection import train_test_split
```

```python
#Usaremos el 80% de los registros para entrenar el modelo y el 20% para validar el modelo
#Hacemos el corte (split)
#x_train será el dataframe pandas que tiene los features X de entrenamiento
#x_test será el dataframe pandas que tiene los features X de validación
#y_train será el dataframe pandas que tiene los labels Y de entrenamiento
#y_test será el dataframe pandas que tiene los labels Y de validación
x_train, x_test, y_train, y_test = train_test_split(dfpFeatures, dfpLabels, test_size=0.2)
```

```python
#Estas cuatro variables son dataframes, por ejemplo:
x_train
```

```python
#Verificamos la dimensionalidad de cada dataframe
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# 6. Definición de la Arquitectura de Red Neuronal Numérica

```python
#Utilitario para definir la arquitectura de una red neuronal
from keras.models import Sequential
```

```python
#Utilitario para definir una capa de red neuronal
from keras.layers import Dense
```

```python
#Instanciamos una arquitectura de red neuronal
modelo = Sequential()
```

```python
#Agregamos una PRIMERA capa de 16 neuronas que procesa 11 features
#Usamos la función de activación "relu" para eliminar la linealidad de las ecuaciones de recta de cada perceptrón
modelo.add(Dense(16, input_shape = (11, ), activation = "relu"))
```

```python
#Agregamos una SEGUNDA capa de 8 neuronas
modelo.add(Dense(8, activation = "relu"))
```

```python
#Agregamos una TERCERA capa de 32 neuronas
modelo.add(Dense(32, activation = "relu"))
```

```python
#Agregamos la CAPA DE SALIDA
modelo.add(Dense(1))
```

# 7. Compilación de la Arquitectura de Red Neuronal

```python
import tensorflow as tf
from keras.optimizers import Adam
```

```python
#Compilamos la arquitectura de red neuronal con el algoritmo e indicamos:
#
# - learning_rate: El ratio de aprendizaje (0.01)
# - "mean_squared_error": La medidad del error en cada iteración (para predicciones numéricas: error cuadrático medio)
#
modelo.compile(Adam(learning_rate = 0.01), "mean_squared_error")
```

# 8. Entrenamiento del modelo

```python
#Entrenamos la arquitectura de red neuronal con 100 iteraciones
modelo.fit(x_train, y_train, epochs = 100, validation_data = (x_test, y_test))
```

# 10. Validación del modelo

```python
#Usamos el dataframe de validación para realizar las predicciones
y_prediccion = modelo.predict(x_test)

#Verificamos
#No se devuelve un dataframe, se devuelve una matriz
y_prediccion
```

```python
#Utilitario para calcular el r2_score
from sklearn.metrics import r2_score
```

```python
#Evaluamos las predicciones
r2_score(y_test, y_prediccion)
```
